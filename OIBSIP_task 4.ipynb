```python
#import required python libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#libraries for NLP
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

# for text processing
import regex
```

```python
import nltk

# Download the stopwords dataset for English
nltk.download("stopwords")
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
sw=stopwords.words('english')
lm=WordNetLemmatizer()
```

```python
#import the dataset
df=pd.read_csv("C:/Users/Sanjiiv30/spam.csv")
df.head()
```

```python
df=df[["v1","v2"]]
```

```python
df.head()
```

```python
df.rename(columns={'v1':'category', 'v2': 'message'}, inplace=True)
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
#Checking for duplicates
df.duplicated().sum()
```

```python
#Checking for null values 
df.isnull().sum()
```

```python
#Dropping Duploicates
df.drop_duplicates()
```

```python
#accessing first 5 rows
df['message'][:5] 
msg=[]
for i in df['message']:
    t=regex.sub('[^A-Za-z0-9]','',i)
    t=t.lower()
    t=word_tokenize(t)
    t=[i for i in t if i not in sw]
    t=[lm.lemmatize(i) for i in t]
    t=" ".join(t)
    msg.append(t)
print(msg)    
```

```python
x=msg
y=df['category']
print(type(x))
print(type(y))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
#Transforming data using Vectoriser
cv=CountVectorizer(max_features=1500)
x_cv=cv.fit_transform(x).toarray()
x_cv.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_cv,y,test_size=0.25)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#building the model
from sklearn.naive_bayes import MultinomialNB
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
def eval_model(y_test,y_pred):
    cm=confusion_matrix(y_test,y_pred)
    print(cm)
    print("Accuracy Score:",accuracy_score(y_test,y_pred))
    print("Classification report:",classification_report(y_test,y_pred))
def model_score(model):
    print("Training score :",model.score(x_train,y_train))
    print("Testing score :",model.score(x_test,y_test))
```

```python
#Predicting values
y_pred_mnb=mnb.predict(x_test)
```

```python
#Finding the model score
model_score(mnb)
```

```python
eval_model(y_test,y_pred_mnb)
```
